```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
trapz_rv = sts.trapz(0.2, 0.8) #объявил случайную величину
```

```python
sample = trapz_rv.rvs(1000) #генерация выборки размера 1000
x = np.arange(0, 1.001, 0.001) #подготовил аргументы
pdf_function_points = trapz_rv.pdf(x) #подготовил точки
plt.plot(x, pdf_function_points, label = "theoretical") #поставил теоретические
plt.hist(sample, normed = True, label = "real", bins = 20) #добавил гистограмму
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
print plt
```

```python
n = [5, 10, 50]
trapz_sred_n = []
for i in n:
     trapz_sred_n.append([sum(trapz_rv.rvs(i)) / i for _ in range(0, 1000)]) #генерирую наборы выборочных средних
```

$E(X) = 0.5$

$D(X) = 0.05(6)$

# Далее идет построение графиков и  гистограмм

```python
sigma = (0.51 / 9 / 5)**0.5 # sqrt(D(X)/n)
mu = 0.5
norm_rv = sts.norm(loc=mu, scale=sigma)
```

```python
plt.title("n = 5")
plt.hist(trapz_sred_n[0], bins = 20, normed = True)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
print plt
```
