````python
import numpy as np
from scipy.stats import norm
from scipy.stats import semicircular
import matplotlib.pyplot as plt
````

# Полукружное распределение

Полукружные случайные величины отличаются от стандартных форм и могут требовать определения дополнительных параметров для завершения спецификации. Параметры: ```x``` : array_like (квантиль), ```q``` : array_like (вероятность восхождения или нисхождения), ```loc``` : array_like, optional (параметр определения местонахождения, задает $\mu$), ```scale``` : array_like, optional (параметр среднеквадратичное отклонение $\sigma$, по умолчанию = 1), ```size``` : int or tuple of ints, optional (форма случайных величин).

````python
# сгенерируем 1000 значений. В данном случае и во всех последующих по умолчанию берутся зачения
# среднего квадратичного отклонения = 0, а среднего мат.ожидания = 0
r = semicircular.rvs(size=1000)
````

````python
# отобразим функцию плотности вероятности (pdf)
x = np.linspace(semicircular.ppf(0.01),
              semicircular.ppf(0.99), 1000)
plt.plot(x, semicircular.pdf(x),
         'r-', lw=5, alpha=0.6, label='semicircular pdf')
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
# отобразим гистограмму
plt.hist(r, density=True, histtype='stepfilled', alpha=0.5)
plt.legend(loc='best', frameon=False)
````

### Оценим распределение выборочного среднего случайной величины при разных объёмах выборок.

Сгенерируем 1000 выборок объема 5 и построим гистограмму распределений их выборочных средних. Поверх гистограммы нарисуем плотность соответствующего нормального распределения, которое рассчитаем используя центральную предельную теорему. 

````python
avg5 = [] # массив для сохранения значений выборочных средних из каждой выборки.
size = 5
sum = 0 # сумма средних, понадобится для расчета значения mu
for i in range(1000):
    r = semicircular.rvs(size=size)
    temp = r.sum()/size
    avg5.append(temp)
    sum += temp
plt.hist(avg5, density=True, histtype='stepfilled', alpha=0.5)

mu = sum / 1000
sigma = 1 / 5. # Согласно ЦПТ для расчета апроксимирующей функции значение дисперсии, 
               # которое по умолчанию задано = 1, делим на объем выборки.

# зададим нормально распределенную случайную величину
norm_rv = norm(loc=mu, scale=sigma)

plt.plot(x, norm_rv.pdf(x),
         'r-', lw=5, alpha=0.6, label='semicircular pdf')
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(loc='best', frameon=False)
````

Сгенерируем 1000 выборок объема 10 и построим гистограмму распределений их выборочных средних. Поверх гистограммы нарисуем плотность соответствующего нормального распределения, которое рассчитаем используя центральную предельную теорему. 

````python
avg10 = [] # массив для сохранения значений выборочных средних из каждой выборки.
size = 10
sum = 0 # сумма средних, понадобится для расчета значения mu
for i in range(1000):
    r = semicircular.rvs(size=size)
    temp = r.sum()/size
    avg10.append(temp)
    sum += temp
plt.hist(avg10, density=True, histtype='stepfilled', alpha=0.5)

mu = sum / 1000
sigma = 1 / 10. # Согласно ЦПТ для расчета апроксимирующей функции значение дисперсии, 
                # которое по умолчанию задано = 1, делим на объем выборки.

# зададим нормально распределенную случайную величину
norm_rv = norm(loc=mu, scale=sigma)

plt.plot(x, norm_rv.pdf(x),
         'r-', lw=5, alpha=0.6, label='semicircular pdf')
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(loc='best', frameon=False)
````

Сгенерируем 1000 выборок объема 50 и построим гистограмму распределений их выборочных средних. Поверх гистограммы нарисуем плотность соответствующего нормального распределения, которое рассчитаем используя центральную предельную теорему.
