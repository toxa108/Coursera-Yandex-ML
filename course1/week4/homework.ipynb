```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math as math
%matplotlib inline

```

#### Строим теоритеческую функцию плотности вероятности и гистограмму по выборке объема 1000 для непрерыного распределения хи квадрат

```python
df = 3
sample = sts.chi2.rvs(df, size=1000)

x = np.linspace(0, 20, 1000)
y = sts.chi2.pdf(x, df)
                 
plt.plot(x, y, 'r-', lw=5, alpha=0.8, label='theoretical pdf')
plt.hist(sample, bins=20, density=True)

plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Гистограмма на основе выборочных средних при n = 5

```python
def createHisto(n, df):
    list_av = list()
    
    for number in range(1000):
        s = sts.chi2.rvs(df, size=n)
        list_av.append(np.mean(s))

    norm_rv = sts.norm(df, math.sqrt(2*df)/math.sqrt(n))
    x = np.linspace(-2,6,100)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf, label='tpdf norm. distr.', alpha=0.8, color='black')
    
    plt.hist(list_av, bins=20, density=True)
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    
createHisto(5, 2)
```

Гистограмма на основе выборочных средних при n = 10

```python
createHisto(10, 2)
```

Гистограмма на основе выборочных средних при n = 50

```python
createHisto(50, 2)
```

Вывод, центральная предельная теорема работает для хи квадрат распределения. Чем выше кол-во элементов в выборке, тем более плотность веротность распределения стремиться к плотности распределения нормального распределения.
