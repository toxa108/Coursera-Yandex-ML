```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from math import sin, exp
from scipy import linalg

def d(x):
    return np.sin(x / 5.) * np.exp(x / 10.) + 5 * np.exp(-x / 2.)

%matplotlib inline
rr = np.arange(1, 15, 0.1)
plt.plot(rr, d(rr).astype(np.float))
plt.title('functio')
plt.show()

x = np.array([[1,1,1,1],[1,4,16,64],[1,10,100,1000],[1,15,225,3375]])
z = np.array([d(1), d(4), d(10), d(15)])
solvee = linalg.solve(x, z)
```
